### Day 11 â€” Data Visualization with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
print("Sample Data:")
print(df.head())
```

##### 1. Distribution Plot (Histogram + KDE)

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['total_bill'], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

##### 2. Boxplot

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['total_bill'])
plt.title("Boxplot of Total Bill")
plt.show()
```

##### 3. Scatter Plot

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex")
plt.title("Total Bill vs Tip")
plt.show()
```

##### 4. Count Plot (Category Frequencies)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="day", data=df)
plt.title("Count of Records by Day")
plt.show()
```

##### 5. Heatmap (Correlation Matrix)

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

##### 6. Pairplot (Multiple Relationships)

```python
sns.pairplot(df, hue="sex")
plt.show()
```

##### 7. Bar Plot

```python
plt.figure(figsize=(6, 4))
sns.barplot(x="day", y="total_bill", data=df, estimator=np.mean)
plt.title("Average Total Bill per Day")
plt.show()
```
