### Day 10 â€” Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

##### 1. Line Plot

```python
x = np.arange(1, 11)
y = x * 2

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y = 2X")
plt.grid(True)
plt.show()
```

##### 2. Bar Chart

```python
categories = ["A", "B", "C", "D"]
values = [10, 25, 15, 30]

plt.figure(figsize=(6, 4))
plt.bar(categories, values)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

##### 3. Histogram

```python
data = np.random.randn(1000)  # normal distribution

plt.figure(figsize=(6, 4))
plt.hist(data, bins=20)
plt.title("Histogram (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

##### 4. Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(6, 4))
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X Random Values")
plt.ylabel("Y Random Values")
plt.show()
```

##### 5. Pie Chart

```python
sizes = [30, 20, 25, 25]
labels = ["Python", "Java", "C++", "JavaScript"]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart of Programming Languages")
plt.show()
```

##### 6. Box Plot

```python
data = [np.random.normal(0, s, 1000) for s in [1, 2, 3]]

plt.figure(figsize=(6, 4))
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Dataset")
plt.ylabel("Values")
plt.show()
```
