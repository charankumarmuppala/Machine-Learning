### Day 5 - Introduction to Pandas (Series & DataFrames)

```python
import pandas as pd
```

##### 1. Creating a Pandas Series

```python
data = [10, 20, 30, 40, 50]
series = pd.Series(data)

print("Pandas Series:")
print(series)
```

```python
# Custom index
series2 = pd.Series([100, 200, 300], index=['a', 'b', 'c'])
print("\nSeries with custom index:")
print(series2)
```

##### 2. Creating a DataFrame

```python
data = {
    'Name': ['Amit', 'Riya', 'John', 'Sara'],
    'Age': [25, 22, 29, 24],
    'City': ['Delhi', 'Mumbai', 'New York', 'London']
}

df = pd.DataFrame(data)
print("\nDataFrame:")
print(df)
```

##### 4. Basic DataFrame Operations

```python
print("\nFirst 2 rows:")
print(df.head(2))
```

```python
print("\nLast 2 rows:")
print(df.tail(2))
```

```python
print("\nSummary info:")
print(df.info())
```

```python
print("\nStatistics:")
print(df.describe())
```

##### 5. Selecting Data

```python
print("\nSelect 'Name' column:")
print(df['Name'])
```

```python
print("\nSelect multiple columns:")
print(df[['Name', 'City']])
```

```python
print("\nSelect row by index (loc):")
print(df.loc[1])
```

```python
print("\nSelect row by position (iloc):")
print(df.iloc[2])
```

##### 6. Adding a new column


```python
df['Score'] = [88, 92, 75, 95]
print("\nDataFrame after adding 'Score':")
print(df)
```

##### 7. Filtering Data

```python
print("\nPeople with Age > 24:")
print(df[df['Age'] > 24])
```
