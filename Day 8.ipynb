### Day 8 â€” Understanding the Dataset (Exploratory Overview)

```python
import pandas as pd
```

```python
data = {
    'Name': ['Amit', 'Riya', 'John', 'Sara', 'Raj'],
    'Age': [25, 22, 29, 24, 30],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'London', 'Chennai'],
    'Salary': [50000, 45000, 60000, 70000, 55000]
}
```

```python
df = pd.DataFrame(data)

print("Dataset:")
print(df)
```

##### 2. Check the shape of the dataset

```python
print("\nShape of dataset (rows, columns):", df.shape)
```

##### 3. Display column names

```python
print("\nColumn Names:", df.columns)
```

##### 4. Data types of each column

```python
print("\nData Types:")
print(df.dtypes)
```

##### 5. Summary statistics for numerical columns

```python
print("\nStatistical Summary:")
print(df.describe())
```

##### 6. Summary statistics for categorical columns

```python
print("\nCategorical Summary:")
print(df.describe(include='object'))
```

##### 7. Display first & last rows

```python
print("\nFirst 3 rows:")
print(df.head(3))
```

```python
print("\nLast 2 rows:")
print(df.tail(2))
```

##### 8. Check unique values

```python
print("\nUnique Cities:", df['City'].unique())
print("Number of Unique Cities:", df['City'].nunique())
```

##### 9. Check for missing values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

##### 10. Basic correlations

```python
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```
